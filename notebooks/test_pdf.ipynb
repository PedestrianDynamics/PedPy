```python
import pathlib
import pandas as pd
import matplotlib.pyplot as plt
import pedpy
```

```python
path = pathlib.Path("demo-data/single_file/")
file = "n56_cam1.csv"
fps = 25
```

```python
df = pd.read_csv(path / file, comment="#")
df = df.rename(columns={"ID": "id", "Frame": "frame", "X": "x", "Y": "y"})[
    ["id", "frame", "x", "y"]
]
traj = pedpy.TrajectoryData(df, fps)
# Compute pair distribution function
radius_bins, pair_distribution = pedpy.compute_pair_distibution_function(traj_data=traj, radius_bin_size=0.1)
```

```python
fig, ax1 = plt.subplots(figsize=(5, 5))
ax1.plot(radius_bins, pair_distribution)
ax1.set_title("Pair Distribution Function")
ax1.set_xlabel("$r$", fontsize=16)
ax1.set_ylabel("$g(r)$", fontsize=16)
ax1.grid(True,alpha=0.3)
plt.show()
```
